# Gyroscope calibration

Basicly we take a look at the LongRun measurement and based on we guess a scale number to normalize the gyroscope data into zirka [-1..1] interwall 

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
lock_meas  = '..\\..\\DATA\\RAW\\Measure_04' # Location folder of the measurement files
meas_date  = '2018_ 11_ 27_9_45_58_'                 # Longest run, complex data set

# Load the IMU csv data file
imu_data = pd.read_csv(lock_meas + '\\LongRun\\' + meas_date + 'IMU.txt',
                               sep='\t',
                               decimal=',',
                               names=['time', 'acc0', 'acc1', 'acc2', 'gyro0', 'gyro1', 'gyro2', 'mag0', 'mag1', 'mag2'])

# Delete invalid measurements
imu_data = imu_data[~(imu_data[['gyro0','gyro1','gyro2']] == 0).any(axis=1)]
# Make array from data frame
imu_ar = np.array(imu_data)

# Create a look up table to see the code more clearly
time=0
gyro0=4
gyro1=5
gyro2=6
```

```python
# Plotting the raw data. Without calibration
plt.plot(imu_ar[:,time], imu_ar[:,gyro0], ' ', linestyle="-", label="gyro0")
plt.plot(imu_ar[:,time], imu_ar[:,gyro1], ' ', linestyle="-", label="gyro1")
plt.plot(imu_ar[:,time], imu_ar[:,gyro2], ' ', linestyle="-", label="gyro2")
plt.legend()
plt.show()
```

```python
scale0 = (imu_ar[:,gyro0].max() - imu_ar[:,gyro0].min()) / 2 
scale1 = (imu_ar[:,gyro1].max() - imu_ar[:,gyro1].min()) / 2 
scale2 = (imu_ar[:,gyro2].max() - imu_ar[:,gyro2].min()) / 2 

print(scale0)
print(scale1)
print(scale2)
```

```python
for i in range(imu_ar.shape[0]):
    imu_ar[i, gyro0] = imu_ar[i, gyro0] / scale0
    imu_ar[i, gyro1] = imu_ar[i, gyro1] / scale1
    imu_ar[i, gyro2] = imu_ar[i, gyro2] / scale2
```

```python
# Plotting the normalized data
plt.plot(imu_ar[:,time], imu_ar[:,gyro0], ' ', linestyle="-", label="gyro0")
plt.plot(imu_ar[:,time], imu_ar[:,gyro1], ' ', linestyle="-", label="gyro1")
plt.plot(imu_ar[:,time], imu_ar[:,gyro2], ' ', linestyle="-", label="gyro2")
plt.legend()
plt.show()
```
